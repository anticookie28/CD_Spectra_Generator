```python
required_packages = ['torch','numpy','scipy','pandas','matplotlib.pylpot']
import warnings
import importlib.util

def check_package(package_name):
    if importlib.util.find_spec(package_name) is None:
        warnings.warn(f'{package_name} is not installed. Please install via typing in your console: pip install {package_name}.')



import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import torch.nn.functional as F
from scipy.signal import find_peaks
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
import os

def readout(fname):
    data_start = False
    data_lines = []
    num_lines_to_read = 300
    
    with open(fname,'r') as file:
        for line in file:
            if "CircularDichroism" in line:
                data_start = True
            if data_start and "Wavelength" in line:
                next(file)
                for _ in range(num_lines_to_read+1):
                    data_lines.append(next(file).strip())
                break

    columns = ["1","2","3","4"]
    data = "\n".join(data_lines)
    df = pd.read_csv(StringIO(data),delim_whitespace=True,names=columns)
    x = df["1"]
    y = 1/3*(df["2"]+df["3"]+df["4"])
    return x,y


def find_spectrum_peaks(fpath,crit='MSELoss',num_epochs=10000,learning_rate=0.01,schedule=True,
               schedule_params = {'patience':50,'factor':0.5},clip=False,
              print_info =True,plotting=True,acc=100,prominence=0.1,improvement_threshhold=1e-10):
    
    print(f'Reading data from: {fpath}')
    x,y = readout(fpath)

    #data = pd.read_csv(fpath, delimiter='\t', header=None, names=['x', 'y'])
    #x = data['x']
    #y = data['y']
    
    class FunctionApproximator(nn.Module):
        def __init__(self):
            super(FunctionApproximator, self).__init__()
            # Define a simple feedforward network
            self.fc1 = nn.Linear(1, 640) 
            self.bn1 = nn.BatchNorm1d(640)
            self.fc2 = nn.Linear(640, 1280) 
            self.bn2 = nn.BatchNorm1d(1280)
            self.fc3 = nn.Linear(1280, 640) 
            self.bn3 = nn.BatchNorm1d(640)
            self.fc4 = nn.Linear(640,320)
            self.bn4 = nn.BatchNorm1d(320)
            self.fc5 = nn.Linear(320,1)

        def forward(self, x):
            x = torch.relu(self.bn1(self.fc1(x)))  # Apply ReLU activation
            x = torch.relu(self.bn2(self.fc2(x)))
            x = F.leaky_relu(self.bn3(self.fc3(x)))
            x = F.leaky_relu(self.bn4(self.fc4(x)))
            x = self.fc5(x)
            return x
    print('Initializing data')    
    x_tensor = torch.tensor(x,dtype=torch.float32).view(-1,1)
    y_tensor = torch.tensor(y,dtype=torch.float32).view(-1,1)
    
    model = FunctionApproximator()
    if crit == 'MSELoss':
        criterion = nn.MSELoss()
    if crit == 'L1Loss':
        criterion = nn.L1Loss()
    if crit == 'SmoothL1Loss':
        criterion = nn.SmoothL1Loss()
        
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    
    if schedule == True:
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min',
                                                         patience=schedule_params['patience']
                                                         ,factor=schedule_params['factor'])
    previous_loss = float('inf')
    print('Start Training')
    for epoch in range(num_epochs):
        model.train()
        optimizer.zero_grad()
        
        outputs = model(x_tensor)
        loss = criterion(outputs, y_tensor)
        
        loss.backward()
        if clip == True:
            nn.utils.clip_norm_grad(model.parameters(),max_norm=1.0)
        
        optimizer.step()
        if schedule == True:
            scheduler.step(loss)
        
        if improvement_threshhold is not None:
            if np.abs(previous_loss - loss.item()) < improvement_threshhold:
                print(f"Stopping training early: Improvement threshhold reached after {epoch} epochs.")
                break
        previous_loss = loss.item()
        
        if print_info == True:
            if (epoch + 1) % 1000== 0:  # Print every 100 epochs
                print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')
    
    print('Evaluating')
    model.eval()
    with torch.no_grad():
        predictions = model(x_tensor)
    if plotting == True:
        print('Plotting for control')
        output_file = fpath.replace('.txt','.png')
        plt.figure(figsize=(12,6))
        plt.plot(x,y,label='measurement data')
        plt.plot(x,predictions,label='predicted data')
        plt.legend()
        plt.xlabel('Wavelength [nm]')
        plt.ylabel('CD [mdeg]')
        plt.savefig(fname=output_file,dpi=600)
        plt.show()
    
    x_test = torch.linspace(max(x),min(x),acc).unsqueeze(1)
    with torch.no_grad():
        predictions2 = model(x_test)                     
    max_loc,_ = find_peaks(predictions2.numpy().reshape(-1,),prominence=prominence)
    min_loc,_ = find_peaks(-predictions2.numpy().reshape(-1,),prominence=prominence)
    
    print('predicted maxima:')
    print(x_test[max_loc])
    print('predicted minima:')
    print(x_test[min_loc])
    
    data_to_save = np.column_stack((np.array(x_test),np.array(predictions2)))
    base_name = os.path.splitext(fpath)[0]
    new_fname = f'{base_name}predictedCD.txt'
    file_path = os.path.join("CDResults",new_fname)
    np.savetxt(file_path,data_to_save,delimiter='\t',fmt='%g',header='WL\tCD')
    print(f'Saving spectrum to {new_fname}')
    
    data_to_save = np.column_stack((np.array(x_test[max_loc]),predictions2[max_loc]))
    base_name = os.path.splitext(fpath)[0]
    new_fname = f'{base_name}predictedmaxima.txt'
    file_path = os.path.join("CDResults",new_fname)
    np.savetxt(file_path,data_to_save,delimiter='\t',fmt='%g',header='WL\tCD')
    print(f'Saving maxima to {new_fname}')
    
    data_to_save = np.column_stack((np.array(x_test[min_loc]),predictions2[min_loc]))
    base_name = os.path.splitext(fpath)[0]
    new_fname = f'{base_name}predictedminima.txt'
    file_path = os.path.join("CDResults",new_fname)
    np.savetxt(file_path,data_to_save,delimiter='\t',fmt='%g',header='WL\tCD')
    print(f'Saving minima to {new_fname}')
    

        
def main(fpath):
    find_spectrum_peaks(fpath)

current_folder = os.getcwd()
txt_files = [f for f in os.listdir(current_folder) if f.endswith('.txt')]

print('Found files')
for index,element in enumerate(txt_files,start=1):
    print(index,'filename:',element)
    


folder_name = "CDResults"
os.makedirs(folder_name,exist_ok=True)



for element in txt_files:
    if __name__ == "__main__":
        try:
            print('     ')
            print('     ')
            file_path = element#input('Please enter the path to the data file:')
            find_spectrum_peaks(fpath=file_path,prominence=1)
            print('        ')
            print('         ')
        except FileNotFoundError:
            print(f'File {file_path} not found.')
```
